```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('results.csv')
df.head(10)
```

# Data Visualization 1: Highest Price of Each Compay During the Day

```python
highest_df = df.groupby('ticker').max()['highest_price'].to_frame()
highest_df.reset_index(inplace=True)
```

```python
ax = sns.catplot(x='ticker',y='highest_price',data=highest_df, kind='bar')
ax.fig.set_size_inches(12,10)
ax.fig.suptitle('Highest Price on December 1st, 2020', fontsize = 16)
```

# Data Visualization 2: Hourly Highest Stock Price Change of Each Company

```python
df = df[['ticker','highest_price','hour']]

for i in df['ticker'].unique():
    plot = df[df['ticker'] == i]
    plot.set_index('hour').plot(kind='line')
    plt.title(f'Hourly Highest Stock Price Change of {i}')
    plt.show()
```
